# Exercise 4
$x = 16*sin(\theta)^3$

$y = 13*cos(\theta) - 5 * cos(2\theta) - 2*cos(3\theta) - cos(4\theta)$

```python
theta = np.linspace(0, 2 * np.pi, 100)
x = 16 * (np.sin(theta) ** 3)
y = 13 * np.cos(theta) - 5 * np.cos(2 * theta)- 2 * np.cos(3 * theta)- np.cos(4 * theta)
plt.plot(x, y)
plt.title('Heart Shape')
plt.show()
```
